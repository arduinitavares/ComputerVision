## Import resources

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
import os
os.chdir(r'C:\Users\atavares\PycharmProjects\ComputerVision')
```

```python
def plot_channels(image):
    channels = {'Red':0, 'Green':1, 'Blue':2}
    fig, axis = plt.subplots(1, 3, figsize=(20, 10))
    for index, value in channels.items():
        channel = image[:, :, value]
        axis[value].set_title(index)
        axis[value].imshow(channel, cmap='gray')
```

```python
# Read in the image
image = cv2.imread('data/images/water_balloons.jpg')

# Make a copy of the image
image_copy = np.copy(image)

# Change color to RGB (from BGR)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

# Display the BGR image
plt.imshow(image_copy)
```

```python
plot_channels(image_copy)
```

```python
# Convert from RGB to HSV
hsv = cv2.cvtColor(image_copy, cv2.COLOR_RGB2HSV)
plot_channels(hsv)
```

## Define pink and hue selection thresholds

```python
# Define our color selection criteria in HSV values
lower_hue = np.array([160,0,0])
upper_hue = np.array([180,255,255])

# Define our color selection criteria in RGB values
lower_pink = np.array([180, 0, 100])
upper_pink = np.array([255, 255, 230])
```

## Mask the image

```python
# RGB mask

# Define the masked area
mask_rgb = cv2.inRange(image_copy, lower_pink, upper_pink)

# Create an copy and mask the image
masked_image = np.copy(image_copy)
masked_image[mask_rgb==0]=[0, 0, 0]

# Display pinks selection
plt.imshow(masked_image)
plt.title('RGB selection')
```

```python
# HSV mask

# Define the masked area
mask_hsv = cv2.inRange(hsv, lower_hue, upper_hue)

# Create an copy and mask the image
masked_image = np.copy(image_copy)
masked_image[mask_hsv==0]=[0, 0, 0]

# Display pinks selection
plt.imshow(masked_image)
plt.title('HSV selection')

```
