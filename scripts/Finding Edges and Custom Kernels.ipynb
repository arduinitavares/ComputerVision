# Creating a Filter, Edge Detection

### Import resources and display image

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
os.chdir(r'C:\Users\atavares\PycharmProjects\ComputerVision')


import cv2
from pprint import pprint
import numpy as np

%matplotlib inline

# Read in the image
image = mpimg.imread('data/images/curved_lane.jpg')

plt.imshow(image)
```

### Convert the image to grayscale


```python
# Convert to grayscale for filtering
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

# Create a custom kernel

```python
# 3x3 array for edge detection
sobel_y_1 = np.array([[-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1]]
)

sobel_y_2 = np.array([[1, 2, 1],
                   [0, 0, 0],
                   [-1, -2, -1]]
)

sobel_x_1 = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]]
)

sobel_x_2 = np.array([[1, 0, -1],
                   [2, 0, -2],
                   [1, 0, -1]]
)

# Create and apply a Sobel x operator
f = np.fft.fft2(gray)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))

# Filter the image using filter 2D, which has inputs:
# grayscale image, bit-depth and kernel
filtered_image_1 = cv2.filter2D(gray, -1, sobel_y_1)
filtered_image_2 = cv2.filter2D(gray, -1, sobel_y_2)
filtered_image_3 = cv2.filter2D(gray, -1, sobel_x_1)
filtered_image_4 = cv2.filter2D(gray, -1, sobel_x_2)

# Display filtered image
fig, axis = plt.subplots(2, 2, figsize=(8, 5))
axis[0, 0].imshow(filtered_image_1, cmap='gray')
axis[0, 1].imshow(filtered_image_2, cmap='gray')
axis[1, 0].imshow(filtered_image_3, cmap='gray')
axis[1, 1].imshow(filtered_image_4, cmap='gray')
```
